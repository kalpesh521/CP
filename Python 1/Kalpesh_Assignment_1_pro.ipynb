## Creating Subsets

```python
data_colm1= data[['Page total likes','Category','Post Month','Paid','share']]
data_colm1
```

```python
#Creating row subset

data_row= data.iloc[[1,4,5,7,9]]
data_row
```

```python
data_row2=data.iloc[[2,8,10]]
data_row2
```

```python
conc= pd.concat([data_row,data_row2])
conc
```

```python
merged = pd.merge(data_row,data_row2,on='Page total likes')
merged

#  merge() is used to combine DataFrames based on a common column or index, performing various types of joins.
# On the other hand, concat() is used to concatenate DataFrames either vertically or horizontally,
# stacking them on top of each other or side by side.
```

```python
#Sort data of subsets 
sort_data= conc.sort_values(by="Category")
sort_data
```

```python
#Transpose dataframe : 
# you can use the transpose() function or the .T attribute to transpose a DataFrame, which essentially flips the rows and columns.
# This operation converts the columns into rows and vice versa

trans= sort_data.transpose()
trans
```

```python
# The resulting melted_df DataFrame is reshaped, 
# where each row represents a unique combination of student name, subject, and score.
# specifies the columns to be melted using the value_vars parameter. the columns to be melted are "Category" and "Post Month". 
# The id_vars parameter is set to "Type" to keep the "Type" column fixed and unchanged

reshape_data= pd.melt(sort_data,id_vars="Type",value_vars=["Category","Post Month"])
reshape_data
```
