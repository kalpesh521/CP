```python
con=pd.concat([df,df3],ignore_index=True)#Concatanation of two dictionaries
con
```

```python
con=pd.concat([df,df3],axis=1,join='inner')#Now we set axes join = inner for intersection of dataframe
con
```

```python
con=pd.concat([df,df3],axis=1,sort=False)#we set axes join = outer for union of dataframe.
con
```

```python
#Group By
grp=con.groupby('Name').groups

grp=con.groupby('Name')
grp.get_group('Sp')
```

```python
#Group By
for Name ,group in grp:
    print(Name)
    print(group)
    print()
```
