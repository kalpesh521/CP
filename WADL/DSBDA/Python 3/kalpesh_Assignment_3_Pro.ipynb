```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv("airquality2.csv")
data
```

```python
data['Ozone'].fillna(data['Ozone'].mean(),inplace = True)
data['Solar.R'].fillna(data['Solar.R'].mean(),inplace=True)
data.drop('Unnamed: 0',axis=1,inplace =True)
```

```python
data['Humidity'].fillna('Medium',inplace=True)
```

```python
data
```

## Piechart

```python
data1 = [data['Ozone'].mean(), data['Solar.R'].mean() ,data['Wind'].mean(), data['Temp'].mean()]
autopct="%1.2f%%"
explode=[0.01, 0.01, 0.01, 0.01]
labels=['Ozone','Solar.R','Wind','Temp']

#data1
plt.pie(x=data1,autopct=autopct,explode=explode,labels=labels)
plt.plot()
```

## Barplot

```python
sns.barplot(data=data,x="Month",y="Temp")
plt.title('Temperature ')
# the black vertical line in the middle of a boxplot represents the median value, 
# providing information about the central tendency of the dataset being visualized
```

```python
plt.bar(data['Month'],data['Temp'])
plt.show()
```

## Box Plot

```python
sns.boxenplot(data=data,x='Month',y='Temp')

'''
Box: The box in the plot represents the interquartile range (IQR), which spans from the 25th percentile (Q1) to the 75th percentile (Q3).
It provides insights into the central tendency and spread of the data.

Median: The line inside the box represents the median value, which is the middle value of the dataset when it is sorted.

Whiskers: The whiskers extend from the box and represent the range of the data. Typically,
they extend to the most extreme data points within a certain distance from the box, usually defined as 1.5 times the IQR.

Outliers: Data points that fall outside the whiskers are considered outliers and are plotted individually as points.
Outliers can indicate unusual or extreme values in the dataset.

Notch: An optional feature of the boxplot is the notch, which is a narrowing of the box around the median.
It provides a rough indication of the confidence interval around the median.
'''
```

```python
from wordcloud import WordCloud , STOPWORDS
# WordCloud is the class that allows us to generate word clouds, and
# STOPWORDS is a predefined set of common words that are often excluded from word clouds.
stopwords=set(STOPWORDS) 
# This line initializes a set of stopwords using the STOPWORDS variable

text='''India is my country Vision is the power that sustains. Whatever we keep in front of our eyes is what will determine our outcome 
 Successful people know that true education is what you get for yourself and by yourself, itâ€™s not what someone gives or
 tells you  they know that true success in life does not come by luck, a lucky man is not a successful man, true winners
 know that they must develop skills and acquire knowledge before becoming truly successful.'''

wordcloud = WordCloud(width=1000,height=1000,background_color='white',
                     stopwords=stopwords,min_font_size=10).generate(text)
#  the generate(text) method generates the word cloud based on the provided input text
plt.imshow(wordcloud) # This line displays the word cloud image using the imshow() function from matplotlib.pyplo
plt.tight_layout(pad=0)# This line adjusts the padding of the plot to ensure all elements are properly displayed.
plt.axis("off") #: This line turns off the axis labels and ticks, so they are not displayed in the plot.
plt.show() #: This line displays the plot with the generated word cloud.
```
