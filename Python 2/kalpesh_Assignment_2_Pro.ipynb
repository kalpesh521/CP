```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
df= pd.read_csv("airquality2.csv")
df
```

```python
df.drop(df.iloc[:,[0]],axis=1,inplace = True)
# inplace=True parameter is commonly used in various methods of data manipulation to modify the object in-place, 
# without creating a new copy of the object. When inplace=True is specified, 
# the method operates directly on the object itself, altering its contents, and returns None.
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Ozone'].fillna(df['Ozone'].mean(),inplace=True)
df['Solar.R'].fillna(df['Solar.R'].mean(),inplace=True)
df['Humidity'].fillna('Medium',inplace=True)
```

```python
df.isnull().sum()
```

```python
'''
inplace=True parameter is commonly used in various methods of data manipulation to modify the object in-place,
without creating a new copy of the object. When inplace=True is specified, the method operates directly on the object itself, 
altering its contents, and returns None.
'''
```

```python
df
```

## Perform label Encoding 

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df['Humidity']=le.fit_transform(df['Humidity'])
```

```python
df
```

```python
x=df[['Ozone']]
y=df[['Temp']]
```

## Perform modeling using Linear regression

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import linear_model
```

```python
regr= linear_model.LinearRegression()
```

```python
model=regr.fit(xtrain,ytrain)
```

```python
predict= model.predict(xtest)
```

```python
predict
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(ytest,predict)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
plt.scatter(xtrain,ytrain,color="red")
plt.plot(xtrain,model.predict(xtrain),color="green")
plt.xlabel('ozone')
plt.ylabel('Temp')
plt.title('Ozone vs Temp ')
```
